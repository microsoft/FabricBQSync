```python
schedule_id = "1234-5678-9012-3456"
project_id = "Project"
dataset = "Dataset"
table_name = "Table"
partition_id = "1234"
config_json_path = "Files/config/fabric-poc-bq-config.json"
```

```python
print(f"Schedule {schedule_id} Loading data for {project_id}.{dataset}.{table_name}${partition_id}...")
```

```python
schedule_loader = BQScheduleLoader(config_json_path)

df = spark.table("LoaderQueue") \
    .filter((col("schedule_id") == lit(schedule_id)) & \
        (col("dataset") == lit(dataset)) & \
        (col("table_name") == lit(table_name)))

if partition_id is not None:
    df = df.filter((col("partition_id") == lit(partition_id)))

for row in df.collect():
    schedule = SyncSchedule(row)

    schedule = schedule_loader.sync_bq_table(schedule)

    schedule_loader.save_schedule_telemetry(schedule) 
```
